```python
#Here is to form the json file

from pathlib import Path
import json
json_path = '/content/drive/MyDrive/YOLO/yolov5/images_rgb_train/coco.json'
path_to_dump = '/content/drive/MyDrive/YOLO/yolov5'
dir_path = Path(path_to_dump)

with open(json_path) as json_file:
    labels_dict = json.load(json_file)

names_list = []
for i in range(len(labels_dict['categories'])):
    names_list.append(' ' + str(labels_dict['categories'][i]['id'] - 1) + ': ' + labels_dict['categories'][i]['name'] + '\n')

file_name = 'flir.yaml'
with open(dir_path.joinpath(file_name),'w') as f:
    #f.write('path: dataset/\n')
    f.write('train: /content/drive/MyDrive/YOLO/yolov5/dataset/train/images\n')
    f.write('val: /content/drive/MyDrive/YOLO/yolov5/dataset/val/images\n')
    f.write('names:\n')
    f.writelines(names_list)
  
! mv /content/drive/MyDrive/YOLO/yolov5/flir.yaml /content/drive/MyDrive/YOLO/yolov5/data
```
